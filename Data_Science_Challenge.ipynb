## Ejercicio 2 con el Dataset Completo

```python
chunksize = 10000
res=[]
for chunk in pd.read_csv('./bookings.csv.bz2', chunksize=chunksize, compression='bz2', delimiter = '^'):
    
    df_save = chunk[['pax','arr_port']].groupby(['arr_port']).agg(['count'])
    df_save.columns = df_save.columns.droplevel()
    res.append(df_save)
```

```python
df_res = res[0]

for i in range(1,len(res)):
    
    df_res = df_res.add(res[i], fill_value=0)
```

```python
df_res = df_res.sort_values('count', ascending=False).head(10)
df_res = df_res['count'].astype(int)
df_res.to_csv('top10', sep='\t')
```
