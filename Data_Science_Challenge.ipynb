### Por chunks

```python
chunksize = 10000
res=[]

for chunk in pd.read_csv('./searches.csv.bz2', chunksize=chunksize, compression='bz2', delimiter = '^', usecols=['Destination', 'Date']):
    
    df_ej3 = chunk[chunk['Destination'].isin(['MAD','BCN','AGP'])]
    df_ej3['Destination'] = pd.Categorical(df_ej3['Destination'], categories=['MAD','BCN','AGP'])
    df_aux=pd.get_dummies(df_ej3['Destination'])
    df_ej3 = df_ej3.join(df_aux)
    df_ej3_plot = df_ej3[['Date','AGP', 'BCN', 'MAD']].copy()
    df_ej3_plot['Date'] = pd.DatetimeIndex(df_ej3_plot.Date).to_period("M")
    df_save = df_ej3_plot.groupby('Date').agg(['sum'])
    df_save.columns = df_save.columns.droplevel()
    df_save.columns = ['AGP','BCN','MAD']
    
    res.append(df_save)
```

```python
df_res = res[0]

for i in range(1,len(res)):
    
    df_res = df_res.add(res[i], fill_value=0)
df_res
```

```python
indexes= df_res.index.values.tolist()
months = []
for i in range(len(indexes)):
    months.append(indexes[i].month)
```

```python
plt.style.use("ggplot")

X = months
Y_AGP = df_res['AGP']
Y_BCN = df_res['BCN']
Y_MAD = df_res['MAD']

plt.plot(X,Y_AGP, label='MÁLAGA')
plt.plot(X,Y_BCN, label='BARCELONA')
plt.plot(X,Y_MAD, label='MADRID')

plt.legend(loc='best', frameon=False)
plt.xlabel

plt.show()
```

```python
any_duplicates = True
res = []
tam = 0

while(any_duplicates):
    
    if(tam==0):
    
        for chunk in pd.read_csv('./searches.csv.bz2', chunksize=chunksize, compression='bz2', delimiter = '^'):

            df_clean = chunk.drop_duplicates()
            res.append(df_clean)


        df_res = res[0]

        for i in range(1,len(res)):

            df_res.append(res[i], ignore_index=True)


        if(df_res.shape[0] == tam):

            any_duplicates = False

        else:

            tam = df_res.shape[0]
            df_res.to_csv('searches_no_duplicates.csv', sep='^')
            
    else:
        
        for chunk in pd.read_csv('./searches_no_duplicates.csv', chunksize=chunksize, delimiter = '^'):

            df_clean = chunk.drop_duplicates()
            res.append(df_clean)


        df_res = res[0]

        for i in range(1,len(res)):

            df_res.append(res[i], ignore_index=True)


        if(df_res.shape[0] == tam):

            any_duplicates = False

        else:

            tam = df_res.shape[0]
            df_res.to_csv('searches_no_duplicates.csv', sep='^')
```
